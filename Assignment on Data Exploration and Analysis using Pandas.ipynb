#### Using the cleaned data you have, carry out exploration and analysis on the data using visualization tools like matplotlib and seaborn where necessary. Remember to write out your findings for each analysis and summarize.


```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# set seaborn style

sb.set(style='whitegrid')
```

```python
# Distribution of total orders over the years

plt.figure(figsize=(10, 6))
sb.boxplot(x='Year', y='Total Orders', data=df)
plt.title('Distribution of Total Orders over the Years')
plt.xlabel('Year')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Monthly trends in Total Sales

plt.figure(figsize=(12, 6))
sb.lineplot(x='Month', y='Total Sales', hue='Year', data=df, marker='o', markersize=8)
plt.title('Monthly Trends in Total Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.show()
```

```python
# Relationship between Gross Sales and Net Sales

plt.figure(figsize=(8, 6))
sb.scatterplot(x='Gross Sales', y='Net Sales', data=df)
plt.title('Relationship between Gross Sales and Net Sales')
plt.xlabel('Gross Sales')
plt.ylabel('Net Sales')
plt.show()
```

```python
# Distribution of Discounts and Returns

plt.figure(figsize=(10, 6))
sb.kdeplot(data=df[['Discounts', 'Returns']], fill=True)
plt.title('Distribution of Discounts and Returns')
plt.xlabel('Amount')
plt.ylabel('Density')
plt.legend(title='Type', labels=['Discounts', 'Returns'])
plt.show()
```

```python
# Yearly comparison of Shipping costs

plt.figure(figsize=(10, 6))
sb.barplot(x='Year', y='Shipping', data=df, estimator=sum)
plt.title('Yearly Comparison of Shipping Costs')
plt.xlabel('Year')
plt.ylabel('Total Shipping Cost')
plt.show()
```

```python
# Comparison of total orders across months

plt.figure(figsize=(12, 6))
sb.barplot(x='Month', y='Total Orders', data=df, palette='viridis')
plt.title('Total Orders Across Months')
plt.xlabel('Month')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.show()
```

### Findings and Summary:

Distribution of Total Orders: The boxplot indicates the distribution of total orders across different years. It provides insights into the variability and spread of order quantities throughout the years.

Monthly Trends in Total Sales: The line plot shows the monthly trends in total sales over the three years. It helps identify seasonal patterns and overall sales trends over time.

Relationship between Gross Sales and Net Sales: The scatter plot illustrates the relationship between gross sales and net sales. It helps visualize how discounts, returns, and other factors affect net sales compared to gross sales.

Distribution of Discounts and Returns: The KDE plot visualizes the distribution of discounts and returns. It provides insights into the frequency and range of discount and return amounts.

Yearly Comparison of Shipping Costs: The bar plot compares the total shipping costs across different years. It helps identify any significant changes or trends in shipping expenses over the years.

Total Orders across Months: The bar plot reveals fluctuations in ordering patterns, with peak months showing higher total orders, suggesting potential seasonal trends in consumer purchasing behavior.
